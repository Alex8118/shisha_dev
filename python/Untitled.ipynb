```python
try:
    from PIL import Image
except ImportError:
    import Image
import pytesseract

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract' 
```

```python
def to_string(link):
    print('image_to_string', pytesseract.image_to_string(link, lang='eng'))
    print('image_to_boxes', pytesseract.image_to_boxes(Image.open(link)))

    img = mpimg.imread(link)
    plt.imshow(img)
    plt.show()
```

```python
to_string(link='620a6d34-82a9-11ea-a416-00155dde8d06-228x228.jpg')
```

```python
import cv2

img_cv = cv2.imread(link)

# By default OpenCV stores images in BGR format and since pytesseract assumes RGB format,
# we need to convert from BGR to RGB format/mode:
img_rgb = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)
print(pytesseract.image_to_string(img_rgb))
# OR
img_rgb = Image.frombytes('RGB', img_cv.shape[:2], img_cv, 'raw', 'BGR', 0, 0)
print(pytesseract.image_to_string(img_rgb))
```
